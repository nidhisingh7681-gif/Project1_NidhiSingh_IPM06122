## Prepare Data for Forecasting Models

### Subtask:
Define 'Global_active_power' as the target variable and split the dataset into training (80%) and testing (20%) sets, ensuring the split maintains the time-series order.


**Reasoning**:
I will define the target variable `y` and the feature matrix `X` from the `df_clustering` DataFrame, then split them into training and testing sets while preserving the time-series order, and finally print their shapes for verification.



```python
import numpy as np

# 1. Define the target variable y as 'Global_active_power'
y = df_clustering['Global_active_power']

# 2. Define the feature matrix X by dropping 'Global_active_power'
X = df_clustering.drop(columns=['Global_active_power'])

# 3. Determine the split point for an 80/20 train-test split based on chronological order
split_point = int(len(df_clustering) * 0.8)

# 4. Split X and y into training and testing sets, maintaining time-series order
X_train = X.iloc[:split_point]
X_test = X.iloc[split_point:]
y_train = y.iloc[:split_point]
y_test = y.iloc[split_point:]

# 5. Print the shapes of X_train, X_test, y_train, and y_test to verify the split
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Train Supervised Forecasting Models

### Subtask:
Train two supervised regression models: Random Forest Regressor and Gradient Boosting Regressor. Train each model in two ways: once on the full training dataset (global model) and separately on subsets of the training data corresponding to each identified cluster (cluster-specific models).


**Reasoning**:
I will import the necessary regressor models, train global models using the full training dataset, and then iterate through each cluster to train cluster-specific models for both Random Forest and Gradient Boosting Regressors, storing them in dictionaries, and finally print a confirmation message.



```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# 2. Initialize and train a RandomForestRegressor on the full training dataset
rf_global_model = RandomForestRegressor(random_state=42)
rf_global_model.fit(X_train, y_train)

# 3. Initialize and train a GradientBoostingRegressor on the full training dataset
gb_global_model = GradientBoostingRegressor(random_state=42)
gb_global_model.fit(X_train, y_train)

print("Global Random Forest and Gradient Boosting models trained.")

# 4. Get the unique cluster labels
unique_clusters = X_train['Cluster'].unique()

# 5. Initialize dictionaries to store cluster-specific models
rf_cluster_models = {}
gb_cluster_models = {}

# 6. Loop through each unique cluster ID to train cluster-specific models
for cluster_id in unique_clusters:
    # a. Filter X_train for the current cluster
    X_train_cluster = X_train[X_train['Cluster'] == cluster_id]
    # b. Filter y_train for the current cluster
    y_train_cluster = y_train.loc[X_train_cluster.index] # Ensure y_train aligns with filtered X_train

    # c. Train a RandomForestRegressor for the current cluster
    rf_model = RandomForestRegressor(random_state=42)
    rf_model.fit(X_train_cluster, y_train_cluster)
    rf_cluster_models[cluster_id] = rf_model

    # d. Train a GradientBoostingRegressor for the current cluster
    gb_model = GradientBoostingRegressor(random_state=42)
    gb_model.fit(X_train_cluster, y_train_cluster)
    gb_cluster_models[cluster_id] = gb_model

print(f"Cluster-specific models trained for {len(unique_clusters)} clusters.")
```

## Model Evaluation and Comparison

### Subtask:
Evaluate the performance of all trained models (global and cluster-specific) using key metrics: Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), and R-squared (R² Score). Present the results in a clear comparison table, discussing whether clustering improves forecasting accuracy.


**Reasoning**:
I need to import evaluation metrics, define a helper function, and then sequentially evaluate all four types of models (global RF, global GB, cluster-specific RF, cluster-specific GB) by making predictions on the test set and calculating RMSE, MAE, and R-squared. Finally, I will compile these results into a comparison table using a pandas DataFrame.



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pandas as pd
import numpy as np

# 1. Define a function to evaluate models
def evaluate_model(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    return rmse, mae, r2

# Initialize a list to store results
results = []

# 2. Evaluate the global Random Forest model
y_pred_rf_global = rf_global_model.predict(X_test)
rmse_rf_global, mae_rf_global, r2_rf_global = evaluate_model(y_test, y_pred_rf_global)
results.append({
    'Model': 'Random Forest (Global)',
    'RMSE': rmse_rf_global,
    'MAE': mae_rf_global,
    'R2 Score': r2_rf_global
})
print(f"Global RF Model - RMSE: {rmse_rf_global:.4f}, MAE: {mae_rf_global:.4f}, R2: {r2_rf_global:.4f}")

# 3. Evaluate the global Gradient Boosting model
y_pred_gb_global = gb_global_model.predict(X_test)
rmse_gb_global, mae_gb_global, r2_gb_global = evaluate_model(y_test, y_pred_gb_global)
results.append({
    'Model': 'Gradient Boosting (Global)',
    'RMSE': rmse_gb_global,
    'MAE': mae_gb_global,
    'R2 Score': r2_gb_global
})
print(f"Global GB Model - RMSE: {rmse_gb_global:.4f}, MAE: {mae_gb_global:.4f}, R2: {r2_gb_global:.4f}")

# 4. Evaluate the cluster-specific Random Forest models
y_pred_rf_cluster = []
y_true_rf_cluster = []

# Ensure X_test indices are aligned with y_test
X_test_aligned = X_test.copy()

for index, row in X_test_aligned.iterrows():
    cluster_id = row['Cluster']
    if cluster_id in rf_cluster_models:
        model = rf_cluster_models[cluster_id]
        # Make sure to pass the features the model was trained on
        # Drop 'Cluster' column as it's not a feature for prediction by the sub-models
        features_for_prediction = [col for col in X_train.columns if col != 'Cluster']
        prediction = model.predict(row[features_for_prediction].to_frame().T)[0]
        y_pred_rf_cluster.append(prediction)
        y_true_rf_cluster.append(y_test.loc[index])
    else:
        # Handle cases where a cluster from test set might not have been in train set (unlikely with dropna, but good practice)
        print(f"Warning: Cluster {cluster_id} not found in trained RF cluster models.")

rmse_rf_cluster, mae_rf_cluster, r2_rf_cluster = evaluate_model(y_true_rf_cluster, y_pred_rf_cluster)
results.append({
    'Model': 'Random Forest (Cluster-Specific)',
    'RMSE': rmse_rf_cluster,
    'MAE': mae_rf_cluster,
    'R2 Score': r2_rf_cluster
})
print(f"Cluster-Specific RF Model - RMSE: {rmse_rf_cluster:.4f}, MAE: {mae_rf_cluster:.4f}, R2: {r2_rf_cluster:.4f}")

# 5. Evaluate the cluster-specific Gradient Boosting models
y_pred_gb_cluster = []
y_true_gb_cluster = []

for index, row in X_test_aligned.iterrows():
    cluster_id = row['Cluster']
    if cluster_id in gb_cluster_models:
        model = gb_cluster_models[cluster_id]
        # Drop 'Cluster' column as it's not a feature for prediction by the sub-models
        features_for_prediction = [col for col in X_train.columns if col != 'Cluster']
        prediction = model.predict(row[features_for_prediction].to_frame().T)[0]
        y_pred_gb_cluster.append(prediction)
        y_true_gb_cluster.append(y_test.loc[index])
    else:
        print(f"Warning: Cluster {cluster_id} not found in trained GB cluster models.")

rmse_gb_cluster, mae_gb_cluster, r2_gb_cluster = evaluate_model(y_true_gb_cluster, y_pred_gb_cluster)
results.append({
    'Model': 'Gradient Boosting (Cluster-Specific)',
    'RMSE': rmse_gb_cluster,
    'MAE': mae_gb_cluster,
    'R2 Score': r2_gb_cluster
})
print(f"Cluster-Specific GB Model - RMSE: {rmse_gb_cluster:.4f}, MAE: {mae_gb_cluster:.4f}, R2: {r2_gb_cluster:.4f}")

# 6. Create a pandas DataFrame to store and display the evaluation metrics
comparison_df = pd.DataFrame(results)
print("\nModel Performance Comparison:")
print(comparison_df.round(4))
```

To fix this, the features provided for prediction to the cluster-specific models must match the features they were trained on, which means including the 'Cluster' column.



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pandas as pd
import numpy as np

# 1. Define a function to evaluate models
def evaluate_model(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    return rmse, mae, r2

# Initialize a list to store results
results = []

# 2. Evaluate the global Random Forest model
y_pred_rf_global = rf_global_model.predict(X_test)
rmse_rf_global, mae_rf_global, r2_rf_global = evaluate_model(y_test, y_pred_rf_global)
results.append({
    'Model': 'Random Forest (Global)',
    'RMSE': rmse_rf_global,
    'MAE': mae_rf_global,
    'R2 Score': r2_rf_global
})
print(f"Global RF Model - RMSE: {rmse_rf_global:.4f}, MAE: {mae_rf_global:.4f}, R2: {r2_rf_global:.4f}")

# 3. Evaluate the global Gradient Boosting model
y_pred_gb_global = gb_global_model.predict(X_test)
rmse_gb_global, mae_gb_global, r2_gb_global = evaluate_model(y_test, y_pred_gb_global)
results.append({
    'Model': 'Gradient Boosting (Global)',
    'RMSE': rmse_gb_global,
    'MAE': mae_gb_global,
    'R2 Score': r2_gb_global
})
print(f"Global GB Model - RMSE: {rmse_gb_global:.4f}, MAE: {mae_gb_global:.4f}, R2: {r2_gb_global:.4f}")

# 4. Evaluate the cluster-specific Random Forest models
y_pred_rf_cluster = []
y_true_rf_cluster = []

# Ensure X_test indices are aligned with y_test
X_test_aligned = X_test.copy()

for index, row in X_test_aligned.iterrows():
    cluster_id = row['Cluster']
    if cluster_id in rf_cluster_models:
        model = rf_cluster_models[cluster_id]
        # The cluster-specific models were trained on X_train_cluster, which included the 'Cluster' column.
        # Therefore, during prediction, we must also provide the 'Cluster' column as a feature to the model.
        features_for_prediction = X_train.columns.tolist()
        prediction = model.predict(row[features_for_prediction].to_frame().T)[0]
        y_pred_rf_cluster.append(prediction)
        y_true_rf_cluster.append(y_test.loc[index])
    else:
        # Handle cases where a cluster from test set might not have been in train set (unlikely with dropna, but good practice)
        print(f"Warning: Cluster {cluster_id} not found in trained RF cluster models.")

rmse_rf_cluster, mae_rf_cluster, r2_rf_cluster = evaluate_model(y_true_rf_cluster, y_pred_rf_cluster)
results.append({
    'Model': 'Random Forest (Cluster-Specific)',
    'RMSE': rmse_rf_cluster,
    'MAE': mae_rf_cluster,
    'R2 Score': r2_rf_cluster
})
print(f"Cluster-Specific RF Model - RMSE: {rmse_rf_cluster:.4f}, MAE: {mae_rf_cluster:.4f}, R2: {r2_rf_cluster:.4f}")

# 5. Evaluate the cluster-specific Gradient Boosting models
y_pred_gb_cluster = []
y_true_gb_cluster = []

for index, row in X_test_aligned.iterrows():
    cluster_id = row['Cluster']
    if cluster_id in gb_cluster_models:
        model = gb_cluster_models[cluster_id]
        # The cluster-specific models were trained on X_train_cluster, which included the 'Cluster' column.
        # Therefore, during prediction, we must also provide the 'Cluster' column as a feature to the model.
        features_for_prediction = X_train.columns.tolist()
        prediction = model.predict(row[features_for_prediction].to_frame().T)[0]
        y_pred_gb_cluster.append(prediction)
        y_true_gb_cluster.append(y_test.loc[index])
    else:
        print(f"Warning: Cluster {cluster_id} not found in trained GB cluster models.")

rmse_gb_cluster, mae_gb_cluster, r2_gb_cluster = evaluate_model(y_true_gb_cluster, y_pred_gb_cluster)
results.append({
    'Model': 'Gradient Boosting (Cluster-Specific)',
    'RMSE': rmse_gb_cluster,
    'MAE': mae_gb_cluster,
    'R2 Score': r2_gb_cluster
})
print(f"Cluster-Specific GB Model - RMSE: {rmse_gb_cluster:.4f}, MAE: {mae_gb_cluster:.4f}, R2: {r2_gb_cluster:.4f}")

# 6. Create a pandas DataFrame to store and display the evaluation metrics
comparison_df = pd.DataFrame(results)
print("\nModel Performance Comparison:")
print(comparison_df.round(4))
```
